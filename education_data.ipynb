### Data cleaning

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(df.isnull())
```

```python
df.isnull().mean()*100
```

```python
dropVar=[]
objVar = []
for var in df.columns:
    if(df[var].isnull().mean()*100>=50):
      dropVar.append(var)
print(dropVar)
```

```python
df2=df.copy()
for var in dropVar:
  df2 = df2.drop(var, axis='columns')
df2
```

```python
df2.info()
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(df2.isnull())
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer = imputer.fit(df2)
  
X = imputer.transform(df2)
df3 = pd.DataFrame(X, columns = df2.columns)
df3
```
